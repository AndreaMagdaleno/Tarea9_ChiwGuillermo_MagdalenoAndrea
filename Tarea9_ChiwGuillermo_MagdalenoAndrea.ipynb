```python
# Primero importamos todas las librerias, paquetes y/o funciones que vamos a utlizar
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definimos los parámetros k, m y c
k, m, c = 3, 1, 0.5

# Función f(y,t) que vamos a integrar
def amortiguado(y,t,k,m,c):
    y1 = y[0]
    y2 = y[1]
    return np.array([y2, -(k/m)*y1-(c/m)*y2])

# Especificamos los puntos de tiempo donde queremos la solución
t = np.linspace(0, 30, 300)
```

### $y_0$ = [1, 1]

```python
# Condiciones iniciales
y0_1 = np.array([1, 1])

# VSolución numérica
y_1 = odeint(amortiguado, y0_1, t, args=(k,m,c))

# Averiguar la forma de solución
y_1.shape

# Extraer y1 y y2
y1 = y_1[:,0]
y2 = y_1[:,1]
```

```python
# Gráfica
omega0=(k/m)**0.5

# Gráfica del plano fase para y0 = [1, 1]
plt.figure(figsize=(8,6))
plt.plot(y1, y2/omega0, 'b', lw=3)
plt.title('Condiciones iniciales: [1, 1]')
plt.xlabel("Posición $y_1$")
plt.ylabel("Velocidad normalizada $y_2/\omega_0$")
plt.grid()
plt.show()
```

### $y_0$ = [1, -1]

```python
# Condiciones iniciales
y0_2 = np.array([1, -1])

# Volvemos a simular con los nuevos parámetros
y_2 = odeint(amortiguado, y0_2, t, args=(k,m,c))

# Averiguar la forma de solución
y_2.shape

# Extraer y1 y y2
y3 = y_2[:,0]
y4 = y_2[:,1]
```

```python
# Gráfica
omega0=(k/m)**0.5

# Gráfica del plano fase para y0 = [1, 1]
plt.figure(figsize=(8,6))
plt.plot(y3, y4/omega0, 'r', lw=3)
plt.title('Condiciones iniciales: [1, 1]')
plt.xlabel("Posición $y_1$")
plt.ylabel("Velocidad normalizada $y_2/\omega_0$")
plt.grid()
plt.show()
```

### $y_0$ = [-1, 1]

```python
# Condiciones iniciales
y0_3 = np.array([-1, 1])

# Volvemos a simular con los nuevos parámetros
y_3 = odeint(amortiguado, y0_3, t, args=(k,m,c))
```

```python
# Gráfica del plano fase para distintos conjuntos de condiciones iniciales
plt.figure(figsize=(10,8))

# Para y0 = [-1, 1]
plt.subplot(2, 2, 3)
plt.plot(y_3[:,0], y_3[:,1], 'g', lw=3)
plt.title('Condiciones iniciales: [-1, 1]')
plt.xlabel('$y_1$ (Posición)')
plt.ylabel('$y_2$ (Velocidad)')
```

### $y_0$ = [-1, -1]

```python
# Condiciones iniciales
y0_4 = np.array([-1, -1])

# Volvemos a simular con los nuevos parámetros
y_4 = odeint(amortiguado, y0_4, t, args=(k,m,c))
```

```python
# Gráfica del plano fase para distintos conjuntos de condiciones iniciales
plt.figure(figsize=(10,8))

# Para y0 = [-1, -1]
plt.subplot(2, 2, 4)
plt.plot(y_4[:,0], y_4[:,1], 'c', lw=3)
plt.title('Condiciones iniciales: [-1, -1]')
plt.xlabel('$y_1$ (Posición)')
plt.ylabel('$y_2$ (Velocidad)')
```
